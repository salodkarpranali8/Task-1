```python
import requests
import json
```

```python
response = requests.get('https://api.openaq.org/v2/cities')
print(response.status_code)
print(response.json()) 
```

## Parms

```python
aqUrl = 'https://api.openaq.org/v2/cities'

aqParams = {'country': 'US', 'count': 2591, 'city': '015'}

aqResp = requests.get(aqUrl, params=aqParams)

print(aqResp.json())
```

## Header

```python
userAgent = 'Mozilla/5.0 (Windows NT 10.0; Win64; x64)'

checkHeader = requests.get('https://api.openaq.org/v2/cities', headers= {'User-Agent' : userAgent})

print(checkHeader)
```
