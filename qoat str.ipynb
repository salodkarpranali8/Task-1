```python
import requests as rq
from bs4 import BeautifulSoup
```

```python
url = 'https://quotes.toscrape.com/'
```

```python
rqst = rq.get(url)
```

```python
rqst
```

```python
print(rqst.content.decode('utf8'))
```

```python
bsoup = BeautifulSoup(rqst._content,'html.parser')
```

```python
bsoup
```

```python
bsoup.title
```

```python
print(bsoup.prettify())
```

```python
qoat = bsoup.find_all("div",class_ ="quote")
```

```python
qoat
```

```python
for i in  qoat:
    print(i.text)
```

```python
qoat_2 = bsoup.find_all("div",class_ ="row")
```

```python
for i in qoat_2:
    print(i.text)
```

```python
qoat_3= bsoup.find_all("div",class_ ="col-md-8")
```

```python
for i in qoat_3:
    print(i.text)
```
